```python
import pandas as pd
```

```python
data = pd.read_csv("D:/Assesments/AI Financial Dashboard/backend/data/extracted_data/Extracted_Data.csv")
data.head()
```

```python
# Remove commas and convert all applicable columns to numeric
numeric_cols = data.columns.drop(["Year", "Top Twenty Shareholders"])
for col in numeric_cols:
    data[col] = data[col].astype(str).str.replace(",", "").str.strip()
    data[col] = pd.to_numeric(data[col], errors="coerce")

data.head()
```

```python
# Parse Top Twenty Shareholders as a list
data["Top Twenty Shareholders"] = data["Top Twenty Shareholders"].apply(
    lambda x: [item.strip() for item in str(x).split(',')] if pd.notnull(x) else []
)
```

```python
data.head()
```

```python
data["Revenue Growth (%)"] = data["Total Revenue"].pct_change() * 100

# Gross Margin (%) = Gross Profit / Total Revenue
data["Gross Margin (%)"] = (data["Gross Profit"] / data["Total Revenue"]) * 100

# Operating Expenses Total
data["Total Operating Expenses"] = (data["Selling and distribution expenses"]
    + data["Administrative expenses"] + data["Other operating expenses"]
)

# Operating Expense Ratio = Total Operating Expenses / Total Revenue
data["Operating Expense Ratio (%)"] = (
    data["Total Operating Expenses"] / data["Total Revenue"]) * 100

# EPS Growth
data["EPS Growth (%)"] = data["Earnings Per Share (EPS)"].pct_change() * 100

# Net Asset Per Share Growth
data["Net Asset Growth (%)"] = data["Net Asset Per Share"].pct_change() * 100
```

```python
def preprocess_financial_data(row):
    
    # Helper function to convert value safely
    def safe_int(value):
        try:
            return int(str(value).replace(',', '').strip()) if pd.notna(value) and str(value).strip() not in ['', 'nan', 'N/A'] else None
        except:
            return None

    def safe_float(value):
        try:
            return float(str(value).strip()) if pd.notna(value) and str(value).strip() not in ['', 'nan', 'N/A'] else None
        except:
            return None

    # Extract top 20 shareholders (assumes they're comma-separated strings)
    top_shareholders_names = [name.strip() for name in str(row.get("Top Twenty Shareholders", "")).split(",")] if pd.notna(row.get("Top Twenty Shareholders")) else []
    
    number_of_shares = [safe_int(share) for share in str(row.get("Number of Shares", "")).split(",")] if pd.notna(row.get("Number of Shares")) else []
    
    share_percentages = [safe_float(pct) for pct in str(row.get("Share %", "")).split(",")] if pd.notna(row.get("Share %")) else []

    return {
        "year": str(row["Year"]).strip() if pd.notna(row["Year"]) else None,
        "industry_group_revenue": {
            "Transportation": safe_int(row["Transportation"]),
            "Consumer Foods": safe_int(row["Consumer Foods"]),
            "Retail": safe_int(row["Retail"]),
            "Leisure": safe_int(row["Leisure"]),
            "Property": safe_int(row["Property"]),
            "Financial Services": safe_int(row["Financial Services"]),
            "Others": safe_int(row["Others"]),
        },
        "total_group_revenue_lkr": safe_int(row["Total Revenue"]),
        "income_statement": {
            "EPS": safe_float(row["Earnings Per Share (EPS)"]),
            "Net Profit": safe_int(row["Net Profit"]),
            "Cost of Sales": safe_int(row["Cost of Sales"]),
            "Gross Profit": safe_int(row["Gross Profit"]),
            "Net Asset Per Share": safe_float(row["Net Asset Per Share"]),
        },
        "operating_expenses_lkr": {
            "Selling and distribution": safe_int(row["Selling and distribution expenses"]),
            "Administrative": safe_int(row["Administrative expenses"]),
            "Other operating": safe_int(row["Other operating expenses"]),
        },
        "top_20_shareholders": {
            "names": top_shareholders_names,
            "shares": number_of_shares,
            "percentages": share_percentages,
        }
    }
```
